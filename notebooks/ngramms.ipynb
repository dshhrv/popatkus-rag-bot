```python
import pandas as pd
import matplotlib.pyplot as plt

ng_3 = pd.read_csv("bmwith.csv")
ng_2 = pd.read_csv("3ngra.csv")



plt.figure()
plt.plot(ng_3["k"], ng_3["recall"], marker="o", label="3 n-граммы")
plt.plot(ng_2["k"],  ng_2["recall"],  marker="o", label="2 n-граммы")
plt.xlabel("K")
plt.ylabel("Recall@K")
plt.grid(True)
plt.legend()
plt.title("Сравнение количества N-грамм")
plt.tight_layout()

plt.savefig("compare_ngramms.png", dpi=200)
plt.show()
```

```python
bm = pd.read_csv("bmwith.csv")
large = pd.read_csv("densebase.csv")



metrics = ["mrr", "hit", "recall"]
titles = ["MRR@K", "Hit@K", "Recall@K"]

fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharex=True)

for ax, metric, title in zip(axs, metrics, titles):
    ax.plot(bm["k"], bm[metric], marker="o", label="sparse (BM25)")
    ax.plot(large["k"], large[metric], marker="o", label="dense")
    ax.set_title(title)
    ax.set_xlabel("K")
    ax.set_ylabel(metric)
    ax.grid(True)

axs[-1].legend(loc="lower right")
plt.tight_layout()
plt.savefig("compare_retrieve.png", dpi=200)

plt.show()
```
