```python
import pandas as pd
import matplotlib.pyplot as plt

rrf = pd.read_csv("datalol.csv")
bm = pd.read_csv("databm.csv")
base = pd.read_csv("densebase.csv")
```

```python
metrics = ["hit", "recall"]
titles = ["Hit@K", "Recall@K"]

fig, axs = plt.subplots(1, 2, figsize=(14, 4), sharex=True)

for ax, metric, title in zip(axs, metrics, titles):
    ax.plot(rrf["k"], rrf[metric], marker="o", label="rrf")
    ax.plot(bm["k"],  bm[metric],  marker="o", label="bm25")
    ax.plot(base["k"], base[metric], marker="o", label="dense")
    ax.set_title(title)
    ax.set_xlabel("K")
    ax.set_ylabel(metric)
    ax.grid(True)

axs[-1].legend(loc="lower right")
# plt.tight_layout()
plt.savefig("compare_rrf2.png", dpi=200)
plt.show()
```
