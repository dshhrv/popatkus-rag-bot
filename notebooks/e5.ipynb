```python
import pandas as pd
import matplotlib.pyplot as plt

large = pd.read_csv("metrics_dense_popatkus-large_dense_2.0_bm25_0.5_rrfk_8_False_2.csv")
base = pd.read_csv("metrics_dense_popatkus-all-small_dense_2.0_bm25_0.5_rrfk_8_False_2.csv")
small = pd.read_csv("metrics_dense_popatkus-base_dense_2.0_bm25_0.5_rrfk_8_False_2.csv")


metric = "recall"

plt.figure()
plt.plot(small["k"], small[metric], marker="o", label="e5-small")
plt.plot(base["k"],  base[metric],  marker="o", label="e5-base")
plt.plot(large["k"], large[metric], marker="o", label="e5-large")

plt.xlabel("K")
plt.ylabel(metric)
plt.grid(True)
plt.legend()
plt.title("Сранение моделей, метрика: recall")
plt.tight_layout()
plt.show()
```

```python
metrics = ["mrr", "ndcg", "recall"]
titles = ["MRR@K", "nDCG@K", "Recall@K"]

fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharex=True)

for ax, metric, title in zip(axs, metrics, titles):
    ax.plot(small["k"], small[metric], marker="o", label="e5-small")
    ax.plot(base["k"],  base[metric],  marker="o", label="e5-base")
    ax.plot(large["k"], large[metric], marker="o", label="e5-large")
    ax.set_title(title)
    ax.set_xlabel("K")
    ax.set_ylabel(metric)
    ax.grid(True)

axs[-1].legend(loc="lower right")
plt.tight_layout()
plt.savefig("compare_dense_3metrics.png", dpi=200)

plt.show()
```
