```python
import pandas as pd
import matplotlib.pyplot as plt
mmarco = pd.read_csv("mmarco.csv")
jina  = pd.read_csv("jina.csv")
rrf = pd.read_csv("datalol.csv")
```

```python
metrics = ["mrr", "ndcg", "recall", "hit"]
titles = ["MRR@K", "nDCG@K", "Recall@K", "Hit@K"]
fig, axs = plt.subplots(1, 4, figsize=(17, 4), sharex=True)

for ax, metric, title in zip(axs, metrics, titles):
    ax.plot(jina["k"], jina[metric], marker="o", label="jina")
    ax.plot(mmarco["k"],  mmarco[metric],  marker="o", label="mmarco")
    ax.plot(rrf["k"],  rrf[metric],  marker="o", label="rrf")
    ax.set_title(title)
    ax.set_xlabel("K")
    ax.set_ylabel(metric)
    ax.grid(True)

axs[-1].legend(loc="lower right")
plt.tight_layout()
plt.savefig("compareencoders.png", dpi=200)
plt.show()
```
