```python
import pandas as pd
import matplotlib.pyplot as plt

ng_3 = pd.read_csv("bmwith.csv")
large = pd.read_csv("densebase.csv")
metric = "recall"

plt.figure()
plt.plot(large["k"], large[metric], marker="o", label="dense")
plt.plot(ng_3["k"],  ng_3[metric],  marker="o", label="sparse")

plt.xlabel("K")
plt.ylabel(metric)
plt.grid(True)
plt.legend(loc="lower right")
plt.title("Сранение методов, метрика: recall")
plt.tight_layout()
plt.show()
```
