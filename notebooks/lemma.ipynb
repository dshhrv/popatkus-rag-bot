```python
import pandas as pd
import matplotlib.pyplot as plt
before = pd.read_csv("bm25_without_lemma.csv")
after  = pd.read_csv("bm25_with_lemma.csv")



metrics = ["hit", "recall", "mrr", "ndcg"]
titles = ["Hit@K", "Recall@K", "MRR@K", "nDCG@K"]

fig, axs = plt.subplots(1, 4, figsize=(15, 4), sharex=True)

for ax, metric, title in zip(axs, metrics, titles):
    ax.plot(before["k"], before[metric], marker="o", label=" ")
    ax.plot(after["k"], after[metric], marker="o", label=" ")
    ax.set_title(title)
    ax.set_xlabel("K")
    ax.set_ylabel(metric)
    ax.grid(True)

axs[-1].legend(loc="lower right")
plt.tight_layout()
plt.savefig("bmcomp.png", dpi=200)

plt.show()
```
